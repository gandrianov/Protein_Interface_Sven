```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors
```

```python
mol_features = []

for mol in Chem.SDMolSupplier("Components-pub.sdf"):
    try:
        name = mol.GetProp("_Name")
        molwt = Descriptors.ExactMolWt(mol)
        num_atom = mol.GetNumAtoms()
        mol_features.append([name, molwt, num_atom])
    except:
        pass
```

```python
data = pd.read_csv('distances.tsv', sep= '\t')
mol_features = pd.DataFrame(mol_features, columns=["Lig", "MW", "NumAtoms"])
```

```python
data["Lig"] = data["Lig:LigChain:ProtChain"].apply(lambda x: x.split(" ")[0])
data["LigChain"] = data["Lig:LigChain:ProtChain"].apply(lambda x: x.split(" ")[1])
data["ProtChain"] = data["Lig:LigChain:ProtChain"].apply(lambda x: x.split(" ")[2])

data = data.merge(mol_features, on="Lig")

data = data[data["Distance"] <= 5.0]
data = data[data["NumAtoms"] >= 15.0]

drop_resns = ["HOH", "ATP" ,"BR","ZN", "NA", "CA", "ADP", "SO4", "PEG","BMA", "EDO", "MSE", "MAN", "ACT", "GSH","UNX","NAG", "MES", "MG", "CL", "PO4","GOL","GTP", "FAD", "GDP", "MN", "DA"]
data = data[~data["Lig"].isin(drop_resns)]
```

```python
ligand_freqs = data.groupby(["Filename", "Lig", "LigChain"], as_index=False).count()
ligand_freqs = ligand_freqs[ligand_freqs["ProtAtomName"] >= 2]

data = data[(data["Filename"].isin(ligand_freqs["Filename"])) &
            (data["Lig"].isin(ligand_freqs["Lig"])) &
            (data["LigChain"].isin(ligand_freqs["LigChain"]))].sort_values(["Filename", "Lig", "LigChain"])
```

```python
ligand_freqs = data.groupby(["Lig", "Filename"], as_index=False).count()[["Lig", "Filename"]]
ligand_freqs = ligand_freqs.groupby(["Lig"], as_index=False).count()
ligand_freqs = ligand_freqs[ligand_freqs["Filename"] == 1]["Lig"]

data = data[data["Lig"].isin(ligand_freqs)].sort_values(["Filename", "Lig", "LigChain"])
data
```
